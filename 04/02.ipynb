```python
# A -> B k1
# B -> C k2
# C -> D k3

# k1 = 1e-3
# k2 = 1e-3
# k3 = 1e-3
# A0 = 0.5 # mol/L
# B0 = 0 # mol/L
# C0 = 0 # mol/L
# D0 = 0 # mol/L










































import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint


def model(y, t):
    A = y[0] # unpack variables
    B = y[1]
    C = y[2]
    D = y[3]
    dAdt = -k1*A
    dBdt = k1*A - k2*B
    dCdt = k2*B - k3*C
    dDdt = k3*C
    return [dAdt, dBdt, dCdt, dDdt]

# Parameters
k1 = 1e-3
k2 = 1e-3
k3 = 1e-3
A0 = 0.5 # mol/L
B0 = 0 # mol/L
C0 = 0 # mol/L
D0 = 0 # mol/L


# Time span
t = np.linspace(0, 3600)
# Initial conditions
ic = [A0, B0, C0, D0]

# Integrate numerically ODEs
results = odeint(model, ic, t)

# Unpack concentrations
A = results[:,0]
B = results[:,1]
C = results[:,2]
D = results[:,3]


plt.plot(t, A, 'b-', label = 'A')
plt.plot(t, B, 'g-', label = 'B')
plt.plot(t, C, 'y-', label = 'C')
plt.plot(t, D, 'r-', label = 'D')
plt.legend()
plt.ylabel('Conc., mol/L')
plt.xlabel('Time, s')
```

```python
# A -> 2B k1
# B -> C k2
# 2C -> D k3

# k1 = 1e-3
# k2 = 1e-3
# k3 = 1e-3
# A0 = 0.5 # mol/L
# B0 = 0 # mol/L
# C0 = 0 # mol/L
# D0 = 0 # mol/L



































import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def model(y, t):
    A = y[0] # unpack variables
    B = y[1]
    C = y[2]
    D = y[3]
    dAdt = -k1*A
    dBdt = 2*k1*A - k2*B
    dCdt = k2*B - 2*k3*C*C
    dDdt = k3*C*C
    return [dAdt, dBdt, dCdt, dDdt]

# Parameters
k1 = 1e-3
k2 = 1e-3
k3 = 1e-3
A0 = 0.5 # mol/L
B0 = 0 # mol/L
C0 = 0 # mol/L
D0 = 0 # mol/L


# Time span
t = np.linspace(0, 3*3600) #20*3600
# Initial conditions
ic = [A0, B0, C0, D0]

# Integrate numerically ODEs
results = odeint(model, ic, t)

# Unpack concentrations
A = results[:,0]
B = results[:,1]
C = results[:,2]
D = results[:,3]


plt.plot(t, A, 'b-', label = 'A')
plt.plot(t, B, 'g-', label = 'B')
plt.plot(t, C, 'y-', label = 'C')
plt.plot(t, D, 'r-', label = 'D')
plt.legend()
plt.ylabel('Conc., mol/L')
plt.xlabel('Time, s')
```
