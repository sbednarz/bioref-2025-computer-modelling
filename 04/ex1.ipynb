```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

# X + S => P

# dX/dt = k*X**1.5



def model(y, t):
    X = y[0]
    dXdt = -k * X**a
    return [dXdt]

k = 5.5e-1
a = 1.5
X0 = 0.02

t = np.linspace(0, 100)
results = odeint( model, [X0], t)
X = results[:,0]
plt.plot(t,X,'r-')
```
