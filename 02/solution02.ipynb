```python
import numpy as np
from scipy.optimize import fsolve

def model(X):
    
    F2 = X[0]
    F3 = X[1]
    x3p = X[2]
    x3b = X[3]
    
    eq1 = F1 - F2 - F3
    eq2 = F1*x1e - F2*x2e
    eq3 = F1*x1p - F2*x2p - F3*x3p
    eq4 = F1*x1b - F3*x3b
    
    return [eq1, eq2, eq3, eq4]


F1 = 100 # mol/h
x1e = 0.333
x1p = 0.333
x1b = 0.333

x2e = 0.666
x2p = 0.333

guess = [1,1,1,1]

F2, F3, x3p, x3b = fsolve(model, guess)

print("solutions: {:.3f} {:.3f} {:.3f} {:.3f}".format(F2, F3, x3p, x3b))
```
