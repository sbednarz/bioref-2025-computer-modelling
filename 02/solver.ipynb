Lets solve system of nonlinear equations:

$3x_1^2 - 2x_2^3 = 25$

$5x_1^3 + 4x_2^2 = 139$

```python
import numpy as np
from scipy.optimize import fsolve

def model(X):
    x1, x2 = X
    eq1 = 3*x1**2 - 2*x2**3 - 25
    eq2 = 5*x1**3 + 4*x2**2 - 139
    return [eq1, eq2]

guess = [-100,-100]
x1, x2 = fsolve(model, guess)

# lazy report:
print(x1, x2)

# pretty report:
# print("guess: {} solutions: {:.3f} {:.3f}".format(guess, x1, x2))
```
